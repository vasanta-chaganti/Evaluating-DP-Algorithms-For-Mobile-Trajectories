```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_input = "/home/vzakhar1/research/KTH-traces/KTH_data/kth_ada_trace/ap_coords/"  #replace with your path if needed
path_output = "/home/vzakhar1/research/KTH-traces/AdaTrace/Outputs/"    #replace with your path if needed
```

```python
def read_file(path):
    k_paths = [] # a list of lists: each list is a path of APs
    with open(path, 'r') as f:
        for line in f:
            if line.startswith('>0:'):
                locations = line.strip('>0:').split(';')
                path = []
                for location in locations[:-1]: 
                    x_co, y_co = location.split(',') # last ele is \n
                    path.append((float(x_co), float(y_co)))
                k_paths.append(path)
    return k_paths
```

```python
def plot_bars(paths_orig, paths_noisy, date, eps):
    k_path_lengths_orig = [len(path) for path in paths_orig]
    k_path_lengths_noisy = [len(path) for path in paths_noisy]
    
    
    plt.hist(k_path_lengths_orig, bins=50, label = 'Original', alpha=0.8, color = 'red')
    plt.hist(k_path_lengths_noisy, bins=50, label = 'Noisy', alpha=0.8)
    plt.title(f'Distribution of Path Lengths in KTH Orig and Noisy Dataset {date} for epsilon {eps}')
    plt.xlabel('Path Length')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

## Noisy path distribution graphing

```python
dates = ["03-03-1", "03-03-2", "04-28-1", "04-28-2", "09-15-1", "09-15-2", "12-08-1", "12-08-2"]
epsilons = [0.1, 1.0, 5.0, 10.0, 20.0, 50.0]
for date in dates:
    for eps in epsilons:
        print("Date: ", date, "\n Epsilon: ", eps)
        input_file = path_input + date + "_output_coordinates.dat"
        output_file = path_output + date + "_output_rounded/" + date + "_eps" + str(eps) + "_rounded.dat"
        paths_input = read_file(input_file)
        paths_noisy = read_file(output_file)
        plot_bars(paths_input, paths_noisy, date, eps)
```
