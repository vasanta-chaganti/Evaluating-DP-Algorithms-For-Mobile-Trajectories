````python
import networkx as nx
import matplotlib.pyplot as plt
import random
import matplotlib.colors as mcolors
````

````python
def read_file(filename):
    graphs = []
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith('>0:'):
                locations = line.strip('>0:').split(';')
                g = nx.empty_graph(0, nx.DiGraph())
                for i in range(len(locations)-2): 
                    x_co, y_co = locations[i].split(',') # last ele is \n
                    x_co_next, y_co_next = locations[i+1].split(',')
                    g.add_edge((float(x_co), float(y_co)),(float(x_co_next), float(y_co_next)) )
                graphs.append(g)
        """
        g = nx.empty_graph(0, nx.DiGraph())
        for n in range(len(traj)-2):
            coords = traj[n].split(",")
            x = coords[0]
            y = coords[1]
            a = (x,y)
            coords_next = traj[n+1].split(",")
            x_next = coords_next[0]
            y_next = coords_next[1]
            b = (x_next, y_next)
            g.add_edge(a,b)
        graphs.append(g)
        """      
    return graphs
````

````python
def draw_with_coords(graphs_orig, graphs_noisy, date, eps):
        #fig, (ax1, ax2) = plt.subplots()
        fig, ax = plt.subplots(1,2, sharey=False,sharex=False, figsize=(20, 10))
        #ax1, ax2 = fig.subplots(2, 1, sharey=False,sharex=False)
        #original graph
        
        G1 = graphs_orig[0]
        nx.set_edge_attributes(G1, 'red', 'color')
        for i in range(1, len(graphs_orig)):
            random_color = random.choice(list(mcolors.CSS4_COLORS.keys()))
            curr = graphs_orig[i]
            nx.set_edge_attributes(curr, random_color, 'color')
            G1 = nx.compose(G1, curr)
        
        positions1 = {}
        for node in list(G1.nodes()):
                if node not in positions1.keys():
                    x = node[0]
                    y = node[1]
                    coords = [float(x), float(y)]
                    positions1[node] = coords

        pos1 = positions1
        edge_colors = nx.get_edge_attributes(G1, 'color')
        plt.subplot(1,2,1)
        nx.draw(G1, pos = pos1, ax = ax[0], connectionstyle='arc3, rad = 0.1', node_size=20, edge_color=list(edge_colors.values()), with_labels=False)
        #nx.draw_networkx_labels(G, pos=pos)
        plt.axis("on")
        ax[0].tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
        plt.xlabel("X-coordinates")
        plt.ylabel('Y-coordinates')

        #noisy graph

        G2 = graphs_noisy[0]
        nx.set_edge_attributes(G2, 'red', 'color')
        for i in range(1, len(graphs_noisy)):
            random_color = random.choice(list(mcolors.CSS4_COLORS.keys()))
            curr = graphs_noisy[i]
            nx.set_edge_attributes(curr, random_color, 'color')
            G2 = nx.compose(G2, curr)
        
        positions2 = {}
        for node in list(G2.nodes()):
                if node not in positions2.keys():
                    x = node[0]
                    y = node[1]
                    coords = [float(x), float(y)]
                    positions2[node] = coords

        pos2 = positions2
        edge_colors = nx.get_edge_attributes(G1, 'color')
        plt.subplot(1,2,2)
        nx.draw(G2, pos = pos2, ax = ax[1], connectionstyle='arc3, rad = 0.1', node_size=20, edge_color=list(edge_colors.values()), with_labels=False)
        #nx.draw_networkx_labels(G, pos=pos)
        plt.axis("on")
        ax[1].tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
        plt.xlabel("X-coordinates")
        plt.ylabel('Y-coordinates')

        #result
        fig.suptitle(f'Original and noisy trajectories on {date} for epsilon {eps}')
        #plt.savefig('foo.png')
        plt.show()
````

````python
def big_func(filename_orig, filename_noisy, date, eps):
    orig = read_file(filename_orig)
    noisy = read_file(filename_noisy)

    draw_with_coords(orig, noisy, date, eps)
````
